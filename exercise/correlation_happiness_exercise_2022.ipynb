```python
df = pd.read_csv('./rankings.csv')
```

```python
df
```

```python
df =  df[['Country', 'RANK','Explained by: Healthy life expectancy']]
df
```

```python
df.rename(columns = {'Explained by: Healthy life expectancy':'exercise'},inplace=True)
df
```

```python
average = []

for x in range(0, 146, 10) :
    start = x
    end  = x + 10
    temp = df[(df['RANK'] >= start) & (df['exercise'] < end)]
    print("{}이상 {}미만 : {}".format(start, end, temp['exercise'].mean()))
    average.append(temp['exercise'].mean())
```
