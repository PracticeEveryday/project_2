```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('mode.chained_assignment',  None) # <==== 경고를 끈다
```

```python
df = pd.read_csv('./2015.csv')
df
```

```python
df =  df[['Happiness Rank','Health (Life Expectancy)']]
df
```

```python
df.rename(columns = {'Health (Life Expectancy)':'Health'},inplace=True)
df
```

```python
average = []

for x in range(0, 157, 10) :
    start = x
    end  = x + 10
    temp = df[(df['Happiness Rank'] >= start) & (df['Health'] < end)]
    print("{}이상 {}미만 : {}".format(start, end, temp['Health'].mean()))
    average.append(temp['Health'].mean())
```

```python
average
```

```python
average
x = np.arange(16)
```

```python
plt.plot(x, average)
```
