```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('drinks.csv')
df
```

```python
beer = df[['country', 'beer_servings']].sort_values('beer_servings', ascending=False).reset_index(drop=True)
beer
```

```python
cocktail = df[['country', 'spirit_servings']].sort_values('spirit_servings', ascending=False).reset_index(drop=True)
cocktail
```

```python
wine = df[['country', 'wine_servings']].sort_values('wine_servings', ascending=False).reset_index(drop=True)
wine
```

```python
beer_top10 = beer.head(10)
beer_top10
```

```python
cocktail_top10 = cocktail.head(10)
cocktail_top10
```

```python
wine_top10 = wine.head(10)
wine_top10
```

```python
beer_top10_average = beer_top10['beer_servings'].mean()
beer_top10_average
```

```python
cocktail_top10_average = cocktail_top10['spirit_servings'].mean()
cocktail_top10_average
```

```python
wine_top10_average = wine_top10['wine_servings'].mean()
wine_top10_average
```
